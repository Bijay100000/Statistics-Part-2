1. What is hypothesis testing in statistics?
ans>> Hypothesis testing is a statistical method used to make decisions or draw conclusions about a population based on sample data. It involves stating two competing hypotheses (null and alternative), selecting a significance level, calculating a test statistic, and deciding whether to reject the null hypothesis based on the result.

2. What is the null hypothesis, and how does it differ from the alternative hypothesis?
ans>> The null hypothesis (Hâ‚€) is a statement that assumes no effect, no difference, or no relationship between variables. It represents the default or status quo assumption.

The alternative hypothesis (Hâ‚ or Ha) is a statement that contradicts the null hypothesis. It suggests there is an effect, a difference, or a relationship.

Difference:
Hâ‚€ (Null Hypothesis): Assumes nothing has changed.
Example form: "There is no difference in means."

Hâ‚ (Alternative Hypothesis): Claims something has changed.
Example form: "There is a difference in means."

In hypothesis testing, we test the null hypothesis and either reject it or fail to reject it based on the evidence from the sample. We never "prove" either hypothesis.

3. What is the significance level in hypothesis testing, and why is it important?
ans>> The **significance level** (denoted as **Î±**) in hypothesis testing is the probability of rejecting the **null hypothesis (Hâ‚€)** when it is actually **true**. It represents the threshold for how much risk of a **Type I error** (false positive) you're willing to accept.

### Common values:

* Î± = 0.05 (5%)
* Î± = 0.01 (1%)
* Î± = 0.10 (10%)

### Why itâ€™s important:

* It sets the **cutoff** for deciding whether the observed data is statistically significant.
* It helps control the likelihood of making incorrect conclusions.
* A lower Î± reduces the chance of a Type I error but increases the chance of a **Type II error** (failing to reject a false null hypothesis).

In short, the significance level determines how strong the evidence must be to reject the null hypothesis.


4. What does a P-value represent in hypothesis testing?
ans>> The **p-value** represents the probability of observing the sample data, or something more extreme, assuming the **null hypothesis** is true.


5. How do you interpret the P-value in hypothesis testing?
ans>> To interpret the **p-value** in hypothesis testing:

* If **p-value â‰¤ significance level (Î±)**: **Reject** the null hypothesis.
* If **p-value > significance level (Î±)**: **Fail to reject** the null hypothesis.

8. What is the Z-test, and when is it used in hypothesis testing?
ans>> The **Z-test** is a statistical test used to determine whether there is a significant difference between sample and population means or between two sample means when the **population variance is known** and the **sample size is large** (usually n â‰¥ 30).

### Used when:

* Population standard deviation is known.
* Sample size is large.
* Data is approximately normally distributed.


9.  How do you calculate the Z-score, and what does it represent in hypothesis testing?
ans>> **Z-score formula** in hypothesis testing:

$$
Z = \frac{\bar{X} - \mu}{\sigma / \sqrt{n}}
$$

Where:

* $\bar{X}$ = sample mean
* $\mu$ = population mean
* $\sigma$ = population standard deviation
* $n$ = sample size

**Represents**:
The **Z-score** shows how many standard errors the sample mean is away from the population mean under the null hypothesis.

16. How is Bayes' Theorem used in statistics, and what is its significance?
ans>> **Bayes' Theorem** is used in statistics to update the **probability of a hypothesis** based on **new evidence** or data.

### Formula:

$$
P(A|B) = \frac{P(B|A) \cdot P(A)}{P(B)}
$$

### Significance:

* Allows **revising probabilities** as more data becomes available.
* Used in **decision making**, **machine learning**, **medical testing**, and **risk assessment**.
* Shifts analysis from **frequency-based** to **probability-based (Bayesian)** reasoning.


17. What is the Chi-square distribution, and when is it used?
ans>> The **Chi-square distribution** is a statistical distribution used to test relationships between categorical variables.

### Used when:

* Performing **Chi-square tests** like:

  * **Goodness-of-fit test** (to see if observed data fits an expected distribution)
  * **Test of independence** (to check if two categorical variables are related)
* Data is in **frequency (count) form**.
* Variables are **categorical** and **independent**.

It is **right-skewed** and depends on the **degrees of freedom**.


18. What is the Chi-square goodness of fit test, and how is it applied?
ans>> The **Chi-square goodness of fit test** is used to determine whether a **sample distribution** matches an **expected (theoretical) distribution**.

### How itâ€™s applied:

1. **State hypotheses**:

   * Hâ‚€: Observed frequencies match expected frequencies.
   * Hâ‚: Observed frequencies do not match expected frequencies.

2. **Calculate test statistic**:

$$
\chi^2 = \sum \frac{(O - E)^2}{E}
$$

Where:

* $O$ = Observed frequency
* $E$ = Expected frequency

3. **Compare** the calculated $\chi^2$ value with the critical value from the Chi-square distribution table at a chosen significance level (Î±) and degrees of freedom.

4. **Decision**:

   * If $\chi^2$ > critical value â†’ **Reject Hâ‚€**.
   * If $\chi^2$ â‰¤ critical value â†’ **Fail to reject Hâ‚€**.


19. What is the F-distribution, and when is it used in hypothesis testing?

### Used in hypothesis testing when:

* Performing an **ANOVA (Analysis of Variance)** to compare **means** of three or more groups.
* Testing if **two population variances** are significantly different.

It is **right-skewed** and depends on **two different degrees of freedom**: one for the numerator and one for the denominator.


20. What is an ANOVA test, and what are its assumptions?
ans>> **ANOVA (Analysis of Variance)** is a statistical test used to determine whether there are **significant differences between the means** of **three or more groups**.

### Assumptions of ANOVA:

1. **Independence** of observations
2. **Normality**: Data in each group should be approximately normally distributed
3. **Homogeneity of variances**: All groups should have equal variances


21. What are the different types of ANOVA tests?
ans>> The **different types of ANOVA tests** are:

1. **One-Way ANOVA**
   â€“ Compares means of **three or more groups** based on **one independent variable**.

2. **Two-Way ANOVA**
   â€“ Compares means based on **two independent variables**.
   â€“ Can test for **interaction effects** between variables.

3. **Repeated Measures ANOVA**
   â€“ Used when the **same subjects** are measured under **different conditions** or over **time**.

4. **MANOVA (Multivariate ANOVA)**
   â€“ Extension of ANOVA when there are **two or more dependent variables**.


22. What is the F-test, and how does it relate to hypothesis testing?
ans>> The **F-test** is a statistical test used to compare **two variances** or to test **overall significance** in models like **ANOVA** or **regression**.

### In hypothesis testing:

* It tests whether **group variances are equal** or whether **group means differ** significantly (in ANOVA).
* The test statistic follows an **F-distribution**.

### Relation to hypothesis testing:

* **Null hypothesis (Hâ‚€)**: No difference in variances or group means.
* **Alternative hypothesis (Hâ‚)**: At least one variance or group mean is different.
* If the **F-value** is greater than the critical value, **reject Hâ‚€**.

```python
# 3.Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution.
#ans>>
def bernoulli_pdf(x, p):
    """
    Calculate the PDF of a Bernoulli distribution.

    Parameters:
    x : int (0 or 1) - outcome
    p : float (0 â‰¤ p â‰¤ 1) - probability of success (x = 1)

    Returns:
    float - probability of the outcome
    """
    if x not in [0, 1]:
        raise ValueError("x must be 0 or 1")
    if not (0 <= p <= 1):
        raise ValueError("p must be between 0 and 1")

    return p**x * (1 - p)**(1 - x)

# ðŸ”½ Call the function and print output
print("P(X=1):", bernoulli_pdf(1, 0.7))  # Output: 0.7
print("P(X=0):", bernoulli_pdf(0, 0.7))  # Output: 0.3
```

```python
#9. Write a Python function to calculate Z-scores from a dataset and plot them.
#ans>>
import numpy as np
import matplotlib.pyplot as plt

def calculate_and_plot_z_scores(data):
    """
    Calculate Z-scores and plot them.

    Parameters:
    data : list or array-like
        Dataset to compute Z-scores from.
    """
    data = np.array(data)
    mean = np.mean(data)
    std = np.std(data)

    if std == 0:
        raise ValueError("Standard deviation is zero, Z-scores cannot be computed.")

    z_scores = (data - mean) / std

    # Plotting Z-scores
    plt.bar(range(len(data)), z_scores, color='skyblue', edgecolor='black')
    plt.axhline(0, color='red', linestyle='--')
    plt.xlabel('Data Index')
    plt.ylabel('Z-score')
    plt.title('Z-scores of the Dataset')
    plt.grid(True)
    plt.show()

    return z_scores

# Example dataset
data = [10, 12, 9, 15, 8, 11, 14]

# Call the function
z_scores = calculate_and_plot_z_scores(data)
print("Z-scores:", z_scores)

```

```python
#15. Implement hypothesis testing using Z-statistics for a sample dataset.
#ans>>
import numpy as np
from scipy.stats import norm

def z_test(sample_data, population_mean, population_std, alpha=0.05):
    """
    Perform a Z-test for a single sample mean.

    Parameters:
    - sample_data: list or array of sample values
    - population_mean: hypothesized population mean (H0)
    - population_std: population standard deviation (assumed known)
    - alpha: significance level (default 0.05)

    Returns:
    - z_score, p_value, conclusion
    """
    sample_data = np.array(sample_data)
    sample_mean = np.mean(sample_data)
    n = len(sample_data)

    # Calculate Z-score
    z_score = (sample_mean - population_mean) / (population_std / np.sqrt(n))

    # Two-tailed p-value
    p_value = 2 * (1 - norm.cdf(abs(z_score)))

    # Decision
    if p_value < alpha:
        conclusion = "Reject the null hypothesis (Hâ‚€)"
    else:
        conclusion = "Fail to reject the null hypothesis (Hâ‚€)"

    return z_score, p_value, conclusion

# ðŸ”¢ Example usage
sample = [105, 110, 98, 100, 102, 108, 95, 99, 101, 103]
population_mean = 100
population_std = 10  # Known population standard deviation

z, p, result = z_test(sample, population_mean, population_std)
print(f"Z-score: {z:.4f}")
print(f"P-value: {p:.4f}")
print("Conclusion:", result)
```

```python
#16. Create a confidence interval for a dataset using Python and interpret the result.
#ans>>
import numpy as np
from scipy import stats

def confidence_interval(data, confidence=0.95):
    """
    Calculate the confidence interval for a dataset using the t-distribution.

    Parameters:
    - data: list or array-like, sample data
    - confidence: confidence level (default = 0.95)

    Returns:
    - (lower_bound, upper_bound)
    """
    data = np.array(data)
    n = len(data)
    mean = np.mean(data)
    std_err = stats.sem(data)  # standard error of the mean

    # t critical value
    t_crit = stats.t.ppf((1 + confidence) / 2, df=n-1)

    margin = t_crit * std_err
    return mean - margin, mean + margin

# ðŸ”¢ Example usage
sample_data = [105, 110, 98, 100, 102, 108, 95, 99, 101, 103]
ci_lower, ci_upper = confidence_interval(sample_data, confidence=0.95)

print(f"95% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")
```

```python
#17. Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean.
#ans>>
import numpy as np
from scipy import stats

# Step 1: Generate data from a normal distribution
np.random.seed(0)  # for reproducibility
data = np.random.normal(loc=50, scale=10, size=100)  # mean=50, std=10, n=100

# Step 2: Calculate 95% confidence interval for the mean
def confidence_interval(data, confidence=0.95):
    n = len(data)
    mean = np.mean(data)
    sem = stats.sem(data)  # standard error of the mean
    t_crit = stats.t.ppf((1 + confidence) / 2, df=n-1)
    margin = t_crit * sem
    return mean - margin, mean + margin

# Compute and print the confidence interval
ci_low, ci_high = confidence_interval(data, confidence=0.95)
print(f"Sample Mean: {np.mean(data):.2f}")
print(f"95% Confidence Interval: ({ci_low:.2f}, {ci_high:.2f})")
```

```python
#1.Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results.
#ans>>
import numpy as np
from scipy.stats import norm

def z_test(sample_data, pop_mean, pop_std, alpha=0.05):
    """
    Perform a Z-test for comparing sample mean to known population mean.

    Parameters:
    sample_data (list or array): Sample data
    pop_mean (float): Known population mean
    pop_std (float): Known population standard deviation
    alpha (float): Significance level (default 0.05)

    Returns:
    None
    """
    sample_mean = np.mean(sample_data)
    n = len(sample_data)

    # Calculate Z-score
    z = (sample_mean - pop_mean) / (pop_std / np.sqrt(n))

    # Calculate p-value for two-tailed test
    p_value = 2 * (1 - norm.cdf(abs(z)))

    print(f"Sample Mean: {sample_mean:.2f}")
    print(f"Z-score: {z:.4f}")
    print(f"P-value: {p_value:.4f}")

    # Interpretation
    if p_value < alpha:
        print("Reject the null hypothesis: Sample mean is significantly different from population mean.")
    else:
        print("Fail to reject the null hypothesis: No significant difference from population mean.")

# Example usage
sample = [51, 49, 52, 50, 48, 47, 53, 50, 49, 52]
population_mean = 50
population_std = 2

z_test(sample, population_mean, population_std)
```

```python
#3.Implement a one-sample Z-test using Python to compare the sample mean with the population mean.
#ans>>
import numpy as np
from scipy.stats import norm

def one_sample_z_test(sample_data, pop_mean, pop_std, alpha=0.05):
    """
    Perform a one-sample Z-test.

    Parameters:
    - sample_data: list or array of sample values
    - pop_mean: known population mean
    - pop_std: known population standard deviation
    - alpha: significance level (default = 0.05)

    Returns:
    - Z-score and p-value
    """
    n = len(sample_data)
    sample_mean = np.mean(sample_data)

    # Calculate Z-score
    z_score = (sample_mean - pop_mean) / (pop_std / np.sqrt(n))

    # Two-tailed p-value
    p_value = 2 * (1 - norm.cdf(abs(z_score)))

    # Print results
    print(f"Sample Mean: {sample_mean:.2f}")
    print(f"Z-score: {z_score:.4f}")
    print(f"P-value: {p_value:.4f}")

    if p_value < alpha:
        print("Reject the null hypothesis: Sample mean is significantly different from population mean.")
    else:
        print("Fail to reject the null hypothesis: No significant difference from population mean.")

    return z_score, p_value

# Example usage
np.random.seed(42)
sample = np.random.normal(loc=52, scale=5, size=30)
population_mean = 50
population_std = 5  # Known population standard deviation

one_sample_z_test(sample, population_mean, population_std)
```

```python
#4.Perform a two-tailed Z-test using Python and visualize the decision region on a plot.
#ans>>
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def two_tailed_z_test_visual(sample_data, pop_mean, pop_std, alpha=0.05):
    """
    Perform a two-tailed Z-test and visualize the decision region.

    Parameters:
    - sample_data: list or array of sample values
    - pop_mean: known population mean
    - pop_std: known population standard deviation
    - alpha: significance level (default = 0.05)
    """
    n = len(sample_data)
    sample_mean = np.mean(sample_data)

    # Calculate Z-score
    z_score = (sample_mean - pop_mean) / (pop_std / np.sqrt(n))

    # Two-tailed p-value
    p_value = 2 * (1 - norm.cdf(abs(z_score)))

    # Critical z-values for two-tailed test
    z_critical = norm.ppf(1 - alpha / 2)

    # Print test results
    print(f"Sample Mean: {sample_mean:.2f}")
    print(f"Z-score: {z_score:.4f}")
    print(f"Critical Z-value: Â±{z_critical:.4f}")
    print(f"P-value: {p_value:.4f}")

    if abs(z_score) > z_critical:
        print("Reject the null hypothesis: Significant difference detected.")
    else:
        print("Fail to reject the null hypothesis: No significant difference.")

    # Plotting the normal distribution with decision regions
    x = np.linspace(-4, 4, 1000)
    y = norm.pdf(x)

    plt.figure(figsize=(10, 5))
    plt.plot(x, y, label='Standard Normal Distribution')

    # Shade critical regions
    plt.fill_between(x, y, where=(x <= -z_critical), color='red', alpha=0.5, label='Rejection Region (Left)')
    plt.fill_between(x, y, where=(x >= z_critical), color='red', alpha=0.5, label='Rejection Region (Right)')

    # Plot Z-score line
    plt.axvline(z_score, color='blue', linestyle='--', label=f'Z-score = {z_score:.2f}')
    plt.axvline(-z_critical, color='black', linestyle='--')
    plt.axvline(z_critical, color='black', linestyle='--')

    plt.title('Two-Tailed Z-Test with Decision Regions')
    plt.xlabel('Z')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True)
    plt.show()

# Example usage
np.random.seed(1)
sample = np.random.normal(loc=52, scale=5, size=30)
population_mean = 50
population_std = 5  # known

two_tailed_z_test_visual(sample, population_mean, population_std)
```

```python
#5.Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.
#ans>>
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def visualize_type1_type2_errors(mu0=50, mu1=52, sigma=5, n=30, alpha=0.05):
    """
    Visualize Type I and Type II errors in hypothesis testing.

    Parameters:
    - mu0 : float : Null hypothesis mean (H0)
    - mu1 : float : Alternative hypothesis mean (H1)
    - sigma : float : Population standard deviation
    - n : int : Sample size
    - alpha : float : Significance level (default = 0.05)
    """

    # Standard error
    se = sigma / np.sqrt(n)

    # Critical value for right-tailed test
    z_crit = norm.ppf(1 - alpha)
    x_crit = mu0 + z_crit * se

    # Type I Error: Rejecting H0 when it's true (area in tail of H0)
    type1_error = alpha

    # Type II Error: Failing to reject H0 when H1 is true
    z_beta = (x_crit - mu1) / se
    beta = norm.cdf(z_beta)

    # Plot distributions
    x = np.linspace(mu0 - 4 * se, mu1 + 4 * se, 1000)
    y0 = norm.pdf(x, mu0, se)  # Under H0
    y1 = norm.pdf(x, mu1, se)  # Under H1

    plt.figure(figsize=(10, 6))
    plt.plot(x, y0, label='Hâ‚€ Distribution (Î¼ = {:.1f})'.format(mu0))
    plt.plot(x, y1, label='Hâ‚ Distribution (Î¼ = {:.1f})'.format(mu1))

    # Shade Type I Error region (Right tail of H0)
    plt.fill_between(x, y0, where=(x >= x_crit), color='red', alpha=0.4, label='Type I Error (Î±)')

    # Shade Type II Error region (Left tail of H1 before x_crit)
    plt.fill_between(x, y1, where=(x < x_crit), color='blue', alpha=0.4, label='Type II Error (Î²)')

    # Decision boundary
    plt.axvline(x_crit, color='black', linestyle='--', label='Critical Value')

    # Labels and legend
    plt.title("Visualization of Type I and Type II Errors")
    plt.xlabel("Sample Mean")
    plt.ylabel("Probability Density")
    plt.legend()
    plt.grid(True)
    plt.show()

    # Print error probabilities
    print(f"Critical sample mean (cutoff): {x_crit:.2f}")
    print(f"Type I Error (Î±): {type1_error:.4f}")
    print(f"Type II Error (Î²): {beta:.4f}")
    print(f"Power of the test (1 - Î²): {1 - beta:.4f}")

# Example usage
visualize_type1_type2_errors(mu0=50, mu1=52, sigma=5, n=30, alpha=0.05)
```

```python
#6.Write a Python program to perform an independent T-test and interpret the results.
#ans>>
import numpy as np
from scipy.stats import ttest_ind

def perform_independent_t_test(sample1, sample2, alpha=0.05):
    """
    Perform an independent two-sample t-test.

    Parameters:
    - sample1 : list or array-like : first sample data
    - sample2 : list or array-like : second sample data
    - alpha : float : significance level (default = 0.05)
    """

    # Perform two-sample t-test
    t_stat, p_value = ttest_ind(sample1, sample2, equal_var=False)  # Welchâ€™s t-test by default

    # Print the test results
    print(f"Sample 1 Mean: {np.mean(sample1):.2f}")
    print(f"Sample 2 Mean: {np.mean(sample2):.2f}")
    print(f"T-statistic: {t_stat:.4f}")
    print(f"P-value: {p_value:.4f}")

    # Interpret the result
    if p_value < alpha:
        print("Reject the null hypothesis: There is a significant difference between the means.")
    else:
        print("Fail to reject the null hypothesis: No significant difference between the means.")

# Example usage
np.random.seed(0)
group1 = np.random.normal(loc=100, scale=10, size=30)
group2 = np.random.normal(loc=105, scale=10, size=30)

perform_independent_t_test(group1, group2)
```

```python
#7.Perform a paired sample T-test using Python and visualize the comparison results.
#ans>>
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import ttest_rel

def paired_t_test_visual(before, after, alpha=0.05):
    """
    Perform paired t-test and visualize the results.

    Parameters:
    - before: list or array of values before treatment
    - after: list or array of values after treatment
    - alpha: significance level (default = 0.05)
    """
    if len(before) != len(after):
        raise ValueError("Both samples must have the same number of observations.")

    # Perform paired t-test
    t_stat, p_value = ttest_rel(before, after)

    # Results
    print(f"Mean (Before): {np.mean(before):.2f}")
    print(f"Mean (After): {np.mean(after):.2f}")
    print(f"T-statistic: {t_stat:.4f}")
    print(f"P-value: {p_value:.4f}")

    # Interpretation
    if p_value < alpha:
        print("Reject the null hypothesis: There is a significant difference between the paired samples.")
    else:
        print("Fail to reject the null hypothesis: No significant difference found.")

    # Visualization
    indices = np.arange(len(before))
    width = 0.35

    plt.figure(figsize=(10, 5))
    plt.bar(indices - width/2, before, width, label='Before', color='skyblue')
    plt.bar(indices + width/2, after, width, label='After', color='lightgreen')
    plt.xlabel('Sample Index')
    plt.ylabel('Values')
    plt.title('Paired Sample Comparison (Before vs. After)')
    plt.legend()
    plt.grid(True, axis='y')
    plt.tight_layout()
    plt.show()

# Example usage
np.random.seed(0)
before = np.random.normal(loc=70, scale=5, size=30)
after = before + np.random.normal(loc=2, scale=3, size=30)  # slight increase after treatment

paired_t_test_visual(before, after)
```

```python
#9.Write a Python function to calculate the confidence interval for a sample mean and explain its significance.
#ans>>
import numpy as np
from scipy.stats import t

def confidence_interval_mean(data, confidence=0.95):
    """
    Calculate the confidence interval for the sample mean.

    Parameters:
    - data : list or array-like : the sample data
    - confidence : float : confidence level (default = 0.95 for 95%)

    Returns:
    - (lower_bound, upper_bound) : tuple of confidence interval
    """
    data = np.array(data)
    n = len(data)
    mean = np.mean(data)
    sem = np.std(data, ddof=1) / np.sqrt(n)  # Standard Error of Mean
    t_critical = t.ppf((1 + confidence) / 2, df=n-1)

    margin_of_error = t_critical * sem
    lower = mean - margin_of_error
    upper = mean + margin_of_error

    return (lower, upper)

# Example usage
np.random.seed(42)
sample = np.random.normal(loc=50, scale=10, size=30)
ci = confidence_interval_mean(sample)
print(f"95% Confidence Interval for the mean: {ci[0]:.2f} to {ci[1]:.2f}")
```

```python
#10.Write a Python program to calculate the margin of error for a given confidence level using sample data.
#ans>>
import numpy as np
from scipy.stats import t

def calculate_margin_of_error(data, confidence=0.95):
    """
    Calculate the margin of error for a sample mean.

    Parameters:
    - data: list or array-like, sample data
    - confidence: float, confidence level (default 0.95)

    Returns:
    - margin_of_error: float
    """
    data = np.array(data)
    n = len(data)
    std_dev = np.std(data, ddof=1)
    sem = std_dev / np.sqrt(n)  # Standard Error of the Mean
    t_critical = t.ppf((1 + confidence) / 2, df=n-1)

    margin_of_error = t_critical * sem
    return margin_of_error

# Example usage
np.random.seed(1)
sample_data = np.random.normal(loc=100, scale=15, size=25)
moe = calculate_margin_of_error(sample_data, confidence=0.95)
print(f"Margin of Error (95% confidence): Â±{moe:.2f}")
```

```python
#11.Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process.
#ans>>
def bayesian_inference(prior, likelihood, false_positive, prevalence_complement):
    """
    Calculates posterior probability using Bayes' Theorem.

    Parameters:
    - prior (P(H)): Prior probability of hypothesis (e.g. disease)
    - likelihood (P(E|H)): Likelihood of evidence given hypothesis
    - false_positive (P(E|~H)): Likelihood of evidence given not hypothesis
    - prevalence_complement (P(~H)): Complement of prior

    Returns:
    - posterior (P(H|E))
    """
    numerator = likelihood * prior
    denominator = numerator + false_positive * prevalence_complement
    posterior = numerator / denominator
    return posterior

# Example: Disease testing
P_D = 0.01  # Prior: Probability of disease
P_Pos_given_D = 0.99  # Likelihood: Positive test if disease
P_Pos_given_not_D = 0.05  # False positive rate
P_not_D = 1 - P_D

posterior = bayesian_inference(P_D, P_Pos_given_D, P_Pos_given_not_D, P_not_D)
print(f"Probability of having the disease given a positive test: {posterior:.4f}")
```
